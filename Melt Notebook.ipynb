# Melting the Dataset

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
dataframe = pd.read_csv("vgsales.csv")
```

## Shaping the Dataframe

For use in Tableau

```python
dataframe
```

```python
dataframemod = dataframe.drop_duplicates()
dataframemod.rename(columns={"NA_Sales": "North America", "EU_Sales": "Europe", "JP_Sales": "Japan", "Other_Sales" : "Other", "Global_Sales" : "Global"}, inplace=True)

sales_columns = ['North America','Europe','Japan', 'Global', 'Other']
melted_dataframe = pd.melt(dataframemod, id_vars=['Name', 'Year', 'Genre', 'Publisher', 'Platform'],
                           value_vars=sales_columns, var_name='Sales_Region', value_name='Sales')


melted_dataframe
```

## Adding Generations

```python
def generation_def(year):
    if year <= 1989:
        return 'First Generation'
    elif 1989 < year <= 1994:
        return 'Second Generation'
    elif 1994 < year <= 2000:
        return 'Third Generation'
    elif 2000 < year <= 2005:
        return 'Fourth Generation'
    elif 2005 < year <= 2017:
        return 'Fifth Generation'
    else:
        return 'Fifth Generation'

# Creating the Column
melted_dataframe['Generation'] = melted_dataframe['Year'].apply(generation_def)

melted_dataframe
```

## Exporting

```python
melted_dataframe.to_csv('vgsalesmod.csv')
```
